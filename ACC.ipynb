```python
import polars as pl
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
file_path="E:/data_analysis/accident"
file_name="US_Accidents.csv"
file=os.path.join(file_path , file_name)
df=pl.read_csv(file)
```

```python
print(f"The head of data: {df.head(5)}")
```

```python
print(df.describe())
```

```python
print("DataFrame shape:" , df.shape)
print("_*_*_*_*_*_*_*_*_*_*_*_*_*_")
print("DataFrame columns:" , df.columns)
print("_*_*_*_*_*_*_*_*_*_*_*_*_*_")
print("DataFrame schema:" , df.schema)
print("_*_*_*_*_*_*_*_*_*_*_*_*_*_")
print("DataFrame Nulls:" , df.null_count())
```

```python
null_count=df.null_count()
missing_report=null_count.unpivot(variable_name="column_name" ,value_name="missing_value_count")
total_rows = df.height

missing_report = missing_report.with_columns([
    ((total_rows - pl.col("missing_value_count")) / total_rows * 100).alias("filling_factor_pct")
])

missing_report = missing_report.sort("filling_factor_pct")
print(missing_report)
```

```python
filtered_report = missing_report.filter(pl.col("filling_factor_pct") < 70)

print(filtered_report)
```

```python
excludes = ["ID","Source","Description","Timezone","Country" ,"End_Lat","End_Lng",
           "Zipcode","Weather_Condition",
           "Airport_Code","Weather_Timestamp",
           "Civil_Twilight","Nautical_Twilight", "Astronomical_Twilight"]
df_cleaned=df.drop(excludes)
print(df_cleaned.head(5))
```

```python
df_cleaned=df_cleaned.unique()
```

```python
print(df_cleaned.head(5))
```

```python
df_size=df_cleaned.estimated_size("gb")
print(df_size)
```

```python
categorical_columns=[col for col, dtype in df_cleaned.schema.items() if dtype == pl.Categorical]
print(f"Categorical columns in dataframe: {df_cleaned.select(categorical_columns)}" , "\n")
String_columns=[col for col, dtype in df_cleaned.schema.items() if dtype == pl.String]
print(f"String columns in dataframe: {df_cleaned.select(String_columns)}")
```

```python
from sklearn.preprocessing import LabelEncoder

labels=LabelEncoder()
encoded = labels.fit_transform(df_cleaned["Sunrise_Sunset"].to_list())

df_cleaned = df_cleaned.with_columns(
    pl.Series("Sunrise_Sunset", encoded)
)

print(labels.inverse_transform([0, 1]) , "\n")
print(df_cleaned.head(5))
```

```python
df_cleaned_small=df_cleaned.with_columns([
    pl.col("Severity").cast(pl.Int8),
    pl.col("Sunrise_Sunset").cast(pl.Int8)]
)
```

```python
df_size2=df_cleaned_small.estimated_size("gb")
print(df_size2)
```

```python
df_parsed=df_cleaned_small.with_columns(
    [
        pl.col("Start_Time").str.to_datetime() ,
        pl.col("End_Time").str.to_datetime()
    ]
).with_columns(
    [
        pl.col("Start_Time").dt.weekday().alias("day_of_week")
    ]
)
df_parsed.head(5)
```

```python

format_days={1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat", 7: "Sun"}
day_count = (
    df_parsed.group_by("day_of_week")    
    .agg(pl.len().alias("count"))         
    .sort("day_of_week")                   
    .with_columns(                         
        pl.col("day_of_week").cast(pl.String).replace(format_days)
    )
)

print(day_count)
```

```python
fig1 = go.Figure(data=[
    go.Bar(
        x=day_count["day_of_week"], 
        y=day_count["count"],
        marker_color='teal'
    )
])
fig1.update_layout(title="Accidents by Day of Week", xaxis_title="Day", yaxis_title="Total Accidents")
fig1.show()
```

```python
time_analysis=(
    df_parsed.with_columns(
    pl.col("Start_Time").dt.hour().alias("Hour")) 
    .group_by(["Hour", "Severity"])
    .agg(pl.len().alias("Severity_Count"))
    .sort("Hour")
)
severity_labels = {
        0: "Low",
        1: "Medium",
        2: "High"
    }
fig2 = go.Figure()
for sev_val, label in severity_labels.items():

    subset = time_analysis.filter(pl.col("Severity") == sev_val)
    
    fig2.add_trace(
            go.Bar(
                x=subset["Hour"],           
                y=subset["Severity_Count"], 
                name=label
            )
        )

fig2.update_layout(
    title="Distribution of Accidents by Hour and Severity",
    xaxis_title="Hour of Day (24H)",    
    yaxis_title="Number of Accidents",  
    barmode="stack", 
    #template="plotly_dark",
    legend_title="Severity Level",
    xaxis=dict(
        tickmode='linear',
        tick0=0,
        dtick=1
    )
)

fig2.show()
```

```python
import plotly.express as px
dn_analysis = (
    df_parsed.with_columns(
        pl.col("Sunrise_Sunset").alias("D/N")
    )
    .group_by(["D/N", "Severity"])
    .agg(pl.len().alias("Severity_Count"))
    .sort("D/N")
)

severity_labels = {
    0: "Low",
    1: "Medium",
    2: "High"
}

colors = px.colors.qualitative.Prism

fig4 = go.Figure()

for i, (sev_val, label) in enumerate(severity_labels.items()):
    subset = dn_analysis.filter(pl.col("Severity") == sev_val)

    fig4.add_trace(
        go.Bar(
            x=subset["D/N"],
            y=subset["Severity_Count"],
            name=label,
            marker_color=colors[i % len(colors)]
        )
    )

fig4.update_layout(
    title="Distribution of Accidents by Daylight and Severity",
    xaxis_title="Day or Night",
    yaxis_title="Number of Accidents",
    barmode="stack",
    legend_title="Severity Level",
    xaxis=dict(
        tickmode="array",
        tickvals=[0, 1],
        ticktext=["day", "night"]
    )
)

fig4.show()
```

```python
state_analysis=(
    df_parsed.with_columns(
    pl.col("State").alias("State")) 
    .group_by(["State", "Severity"])
    .agg(pl.len().alias("Severity_Count"))
    .sort("State")
)
severity_labels = {
        0: "Low",
        1: "Medium",
        2: "High"
    }
fig3 = go.Figure()
for sev_val, label in severity_labels.items():

    subset = state_analysis.filter(pl.col("Severity") == sev_val)
    
    fig3.add_trace(
            go.Bar(
                x=subset["State"],           
                y=subset["Severity_Count"], 
                name=label
            )
        )

fig3.update_layout(
    title="Distribution of Accidents by State and Severity",
    xaxis_title="States",    
    yaxis_title="Number of Accidents",  
    barmode="stack", 
    #template="plotly_dark",
    legend_title="Severity Level",
    xaxis=dict(
        tickmode='linear',
        tick0=0,
        dtick=1
    )
)

fig3.show()
```

```python
cols_to_analyze = [
    "Amenity", "Bump", "Crossing", "Give_Way", "Junction", 
    "No_Exit", "Railway", "Roundabout", "Station", "Stop", 
    "Traffic_Calming", "Traffic_Signal", "Turning_Loop"
]

states = ["CA", "FL", "TX", "SC", "NY"]

fig = make_subplots(
    rows=5, cols=1, 
    shared_xaxes=True, 
    vertical_spacing=0.05,
    subplot_titles=states
)

for idx, state in enumerate(states):
    df_state = df_parsed.filter(pl.col("State") == state).select(cols_to_analyze)
    
    if df_state.height == 0:
        continue

    analysis_data = []
    for col in cols_to_analyze:
        occurrence_rate = df_state[col].mean() * 100
        analysis_data.append({
            "Feature": col.replace("_", " ").title(),
            "Rate": occurrence_rate
        })
    
    summary_df = pl.DataFrame(analysis_data).sort("Rate")
    

    rates = summary_df["Rate"].to_list()
    features = summary_df["Feature"].to_list()
    labels = summary_df["Rate"].map_elements(lambda x: f"{x:.1f}%", return_dtype=pl.Utf8).to_list()

    fig.add_trace(
        go.Bar(
            x=rates,
            y=features,
            orientation='h',
            text=labels,
            textposition='outside',
            marker=dict(color=rates, colorscale='Viridis'),
            showlegend=False
        ),
        row=idx+1, col=1
    )

fig.update_layout(
    height=1500, 
    width=1000, 
    title_text="Traffic Feature Prevalence by State",
    template="plotly_white",
    showlegend=False
)

fig.update_xaxes(title_text="Percentage (%)")
fig.show()
```
